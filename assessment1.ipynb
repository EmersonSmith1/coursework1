Coding Task and Question: By looking back at the tutorial on data cleaning, choose a method for cleaning these dirty data values and implement it (you should use a different approach for each). In the next three blocks:
- include your code for this (about 10 lines)
- explain your approaches
- re-plot the chart from earlier to check that your approach has been effective

```python
sales1 = pd.DataFrame(sales)

#Replaces missing values in SKU14256 with 0
sales1.SKU14256 = round(sales.SKU14256.replace(0, sales.SKU14256.mean()))

#Replaces the highest value in SKU15276 with the median
#The median only accounts uses up to the 24th as the trend changes after then 
#(this has a better result than looking at the median for all data):

sales1.SKU15276 = round(sales.SKU15276.replace(sales.SKU15276.max(), sales.SKU15276[0:24].median()))
sales1
```

- The dirty value in SKU14256 is 0 so I replaced it with the mean so it fits with the general trend. Judging by the replotted graph below, it looks like this was successful.
- The dirty value in SKU15276 was too large so I replaced it with the median to try to make it closer to the trend of the other data points in the line. However, this doesn't work as the trend of sales changes to be much lower after the 24th. I then tried changing it so it calculated the median if the data between the 1st and 24th inclusively and this made the graph a lot better as it doesn't have a huge change on that day.

```python
#Plots a graph of the data without the dirty data
for x in range(1, len(sales.columns)):
    plt.plot(sales1.Date, sales1[sales1.columns[x]], label = sales1.columns[x])

plt.legend()
plt.show()
```
